```python
df = pd.read_csv('sobcov\sobcov89.txt',sep="|",header=None)

colnames=['Year','State.Code','State.Abbr','County.Code','County.Name','Commodity.Code','Commodity.Name', \
                   'Insurance.Plan.Code','Insurance.Plan.Abbr','Coverage.Category','Delivery.Type','Coverage.Level','Policies.Sold.Count','Policies.Earning.Premium', \
                   'Policies.Indemnified','Units.Earning.Premium','Units.Indemnified','Quantity.Type','Net.Planted.Quantity','Net.Endorsed.Acres', 'Liability','Total.Premium',\
                  'Subsidy','State/Provate.Subsidy','Additional.Subsidy','EFA.Premium.Discount','Indemnity.Amount','Loss.Ratio']             
df.head()  
```
